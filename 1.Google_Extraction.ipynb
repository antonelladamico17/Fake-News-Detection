## **1. Google Extraction**

This notebook was used to extract news from Google Serpe and Google
News.\
The keywords on which the searches were made are contained in the list named ```queries```.\
The extraction was performed in incognito mode. From each news link, title and description (or snippet) have been extracted.\
Information retrieved are seved to an excel file. Then, after this extraction process, we proceeded to the manual labeling of each news in order to continue with our experiment.

````python
# Import libraries

from selenium import webdriver
import pandas as pd
from time import sleep
from bs4 import BeautifulSoup
from bs4.element import Tag
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.common.exceptions import NoSuchElementException
````

````python
options = webdriver.ChromeOptions()
options.add_argument('--incognito')  # incognito window
options.add_argument('--lang = en') # set lang to EN to get only eng result
options.add_argument('--disable-cache')
````

````python
# Lists for saving results
links = []
titles = []
descriptions = []
````

````python
# keywords used to extract the results
queries = ['china+new+disease', 'china+virus', 'coronavirus', 'event+gates+201', 'covid19+5g', 'vaccine+coronavirus' ]
````

````python
# Section News
for i in queries:
    driver = webdriver.Chrome(r"C:\Users\Utente\Desktop\chromedriver.exe", options = options)
    driver.get('https://google.com/search?q=' + i + "&lr=lang_en"  + '&tbs=cdr:1,cd_min:12/31/2019,cd_max:03/31/2020' )

    driver.find_element_by_id('L2AGLb').click()
    driver.find_element_by_xpath('//*[@id="hdtb-msb"]/div[1]/div/div[2]/a').click()


    while True:
        try:
            content = driver.page_source
            soup = BeautifulSoup(content, "html.parser")
            driver.find_element_by_xpath('//*[@id="pnnext"]').click()
            sleep(15)
            result_div = None
            result_div = soup.find_all('g-card', attrs={'class': 'ftSUBd'})
            for r in result_div:
                try:
                    link = r.find('a', href=True)
                    title = None
                    title = r.find('div', attrs={'class': 'mCBkyc y355M JQe2Ld nDgy9d'})

                    if isinstance(title, Tag):
                        title = title.get_text()

                    description = None
                    description = r.find('div', attrs={'class': 'GI74Re nDgy9d'})

                    if isinstance(description, Tag):
                        description = description.get_text()

                    if link != '' and title != '' and description != '':
                        links.append(link['href'])
                        titles.append(title)
                        descriptions.append(description)

                except Exception as e:
                    print(e)
                    continue

        except NoSuchElementException:
            break
    driver.close()


df = pd.DataFrame({'Title': titles, 'Link': links, 'Description': descriptions})

file_name = 'google_news_results.xlsx'

df.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
````

````python
# Main page
for i in queries:
    driver = webdriver.Chrome(r"C:\Users\Utente\Desktop\chromedriver.exe", options = options)
    driver.get('http://google.com/search?q=' + i + "&lr=lang_en" + '&tbs=cdr:1,cd_min:12/31/2019,cd_max:07/28/2020')

    driver.find_element_by_id('L2AGLb').click()

    while True:
        try:
            content = driver.page_source
            soup = BeautifulSoup(content, "html.parser")
            driver.find_element_by_xpath('//*[@id="pnnext"]').click()
            sleep(15)
            result_div = None
            result_div = soup.find_all('div', attrs={'class': 'jtfYYd'})
            for r in result_div:
                try:
                    link = r.find('a', href=True)
                    title = None
                    title = r.find('h3', attrs={'class': 'LC20lb MBeuO DKV0Md'})

                    if isinstance(title, Tag):
                        title = title.get_text()

                    description = None
                    description = r.find('div', attrs={'class': 'NJo7tc Z26q7c uUuwM'})

                    if isinstance(description, Tag):
                        description = description.get_text()

                    if link != '' and title != '' and description != '':
                        links.append(link['href'])
                        titles.append(title)
                        descriptions.append(description)

                except Exception as e:
                    print(e)
                    continue

        except NoSuchElementException:
            break
    driver.close()


df = pd.DataFrame({'Title': titles, 'Link': links, 'Description': descriptions})

file_name = 'google_main_results.xlsx'

df.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
````
